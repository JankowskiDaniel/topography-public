```python
import sys
sys.path.insert(0, '/home/shapke/Pulpit/topography-public/')
from src.data_preparation.preparation_utils import prepareDataFrame

chosen_dataset_kinds = ['all']
csv_filenames = list(map(lambda x: f'/home/shapke/Pulpit/{x}_10k.csv', chosen_dataset_kinds))
dataf = prepareDataFrame(csv_filenames, images_loc)
dataf
```

```python
import numpy as np
from PIL import Image
from src.methods.analytic.exact import AnalyticalMathodOld
from tqdm import tqdm

pred_epsilons = []
method = AnalyticalMathodOld()
for indax, row in tqdm(dataf.iterrows(), total=len(dataf)):
    image = Image.open(row.fullpath)
    image = np.array(image)
    
    eps = method.calculate_epsilon(image, check=[], blur_size=10, vector_filter_size=20)
    pred_epsilons.append(*eps)

dataf['pred_epsilon_exact'] = pred_epsilons
dataf
```

```python
def cyclic_MAE(output, target):
    return min(abs(output-target), min(abs(output-target+1), abs(output-target-1)))

dataf['cyclic_error_exact'] = dataf.apply(lambda row: cyclic_MAE(row['epsilon'], row['pred_epsilon_exact']), axis=1)
dataf
```

```python
exact_MAE = dataf['cyclic_error'].mean()
exact_MAE
```
