# The Newest model trained of shadowed parlin noise images for 30 epochs #

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

steel = pd.read_csv('../../data/results/perlin/pizza/best_steel_model.csv')
exact_method = pd.read_csv('../../data/results/exact_method/exact_method-raw_steel_predictions.csv')
```

```python
# split the dataframes into the correct and incorrect files
correct_steel = steel[~steel['filename'].isin(exclude_files)]
incorrect_steel = steel[steel['filename'].isin(exclude_files)]
correct_exact_method = exact_method[~exact_method['filename'].isin(exclude_files)]
incorrect_exact_method = exact_method[exact_method['filename'].isin(exclude_files)]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot for incorrect_steel
axes[0].hist(incorrect_steel['epsilon'], bins=10, range=(0, 1), color='red', alpha=0.5, edgecolor='black')
axes[0].set_title('NN Model: Incorrect images predictions (Steel)')
axes[0].set_xlabel('Epsilon')
axes[0].set_ylabel('Frequency')
axes[0].set_xticks(np.arange(0, 1.1, 0.1))

# Plot for incorrect_exact_method
axes[1].hist(incorrect_exact_method['epsilon'], bins=10, range=(0, 1), color='red', alpha=0.5, edgecolor='black')
axes[1].set_title('Exact Method: Incorrect images predictions (Exact Method)')
axes[1].set_xlabel('Epsilon')
axes[1].set_ylabel('Frequency')
axes[1].set_xticks(np.arange(0, 1.1, 0.1))

plt.tight_layout()
plt.show()
```
