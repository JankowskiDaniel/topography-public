Add triangle noise to the perlin images

```python
import cv2
import pandas as pd
import sys

sys.path.append("../..")
```

```python
from src.data_generation.noise_controllers.noise_pizza import PizzaController
```

```python
controller = PizzaController(
    **{
        "nr_of_pizzas": (3, 5),
        "center_point": (320, 240),
        "channels": 1,
        "strength": (20, 30),
    }
)
```

```python
main_path = "../../data/generated/perlin/"
df1 = pd.read_csv(main_path + "parameters.csv")
df2 = pd.read_csv(main_path + "parameters_2.csv")

# concatenate the two dataframes
df = pd.concat([df1, df2], ignore_index=True)
df.to_csv(main_path + "parameters.csv", index=False)
```

```python
path_in = "../../data/generated/perlin/"
path_out = "../../data/generated/perlin/pizza/"
```

```python
# read image
from tqdm import tqdm

df_pizza = pd.DataFrame()
pizza_data = []
for i, row in tqdm(df.iterrows()):
    row_copy = row.copy()
    img = cv2.imread(path_in + row["filename"], cv2.IMREAD_GRAYSCALE)
    img = controller.generate(img, row["epsilon"])
    new_filename = "pizza_" + row["filename"]
    cv2.imwrite(path_out + new_filename, img)
    row_copy["filename"] = new_filename
    pizza_data.append(row_copy)

df_pizza = pd.DataFrame(pizza_data)
df_pizza.to_csv(path_out + "parameters.csv", index=False)
```
