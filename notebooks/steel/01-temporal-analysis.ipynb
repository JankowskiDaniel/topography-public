```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("../../data/results/perlin/best_steel_model.csv")

def plot_distribution(ax, df: pd.DataFrame, title: str) -> None:
    ax.hist(df['epsilon'], bins=10, range=(0, 1), color='red', alpha=0.5, edgecolor='black')
    ax.set_title(title)
    ax.set_xlabel('Epsilon')
    ax.set_ylabel('Frequency')
    ax.set_xticks(np.arange(0, 1.1, 0.1))

# Example dataframe and epsilon column (for demonstration)
# df = pd.DataFrame({'epsilon': np.random.rand(100000)})

dfs = np.array_split(df, 10)
ranges = ["00000 - 09999", "10000 - 19999", "20000 - 29999", "30000 - 39999", 
          "40000 - 49999", "50000 - 59999", "60000 - 69999", "70000 - 79999", 
          "80000 - 89999", "90000 - 99999"]

# Create one figure with 10 plots (5 columns, 2 rows)
fig, axs = plt.subplots(2, 5, figsize=(20, 10))  # 5 columns, 2 rows
fig.suptitle("Epsilon distribution per 10,000 iterations")

# Flatten axs to make it easier to iterate over
axs = axs.ravel()

for i, df_split in enumerate(dfs):
    plot_distribution(axs[i], df_split, ranges[i])

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust rect to account for suptitle
plt.show()
```
