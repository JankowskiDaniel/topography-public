```python
from glob import glob
import sys
sys.path.append('../src')
```

```python
from methods.analytic.exact import AnalyticalMathodOld
method = AnalyticalMathodOld()
```

```python
NOISE_PATH = "../data/fourier_noise/steel/"
RAW_PATH = "../data/raw/steel/"
```

```python
noise_paths = glob(NOISE_PATH+"*")
raw_indexes = [x.split("_")[-1] for x in noise_paths]
raw_paths = [RAW_PATH+x for x in raw_indexes]
raw_paths[:5]
```

```python
results = []
from tqdm import tqdm
import cv2
for path in tqdm(raw_paths):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    epsilon = method.calculate_epsilon(img)
    results.append(epsilon[0])
```

```python
import seaborn as sns
```

```python
sns.histplot(results, bins=20)\
    .set_title("Epsilon distribution")
```
